```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
#Connecting to webdriver
driver=webdriver.Chrome(r"C:\Users\donga\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
#Accesing the website
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos ")
driver.maximize_window()
time.sleep(2)
```

```python
rank=[]
name=[]
artist=[]
upload_date=[]
views=[]
```

```python
#Getting the rank
number=driver.find_elements(By.XPATH,'//a[@class="mw-redirect" ] ')
for i in number:
    rank.append(i.text)

#Getting the name of video
video_names=driver.find_elements(By.XPATH,'//a  [@class="mw-redirect" ] ')
for video in video_names:
    name.append(video.text)
    
#Getting the artist details
uploaders=driver.find_elements(By.XPATH,'//a [@class="mw-redirect" ] ')
for uploader in uploaders:
    artist.append(uploader.text)
    
#Getting the upload date
uploaded_on=driver.find_elements(By.XPATH,' //a [@class="mw-redirect" ]')
for uploaded in uploaded_on:
    upload_date.append(uploaded.text)

#Getting the views
tot_views=driver.find_elements(By.XPATH,'//a [@ class="mw-redirect" ] ')
for view in tot_views:
    views.append(view.text)
    

driver.close()

```

```python
youtube=pd.DataFrame({'Rank':rank,'Name':name,'Artist':artist,'Uploaded On':upload_date, 'Views in billions': views})
youtube
```

Q10- Details of Datasets from UCI machine learning repositories.

```python
#Connecting to webdriver
driver=webdriver.Chrome(r"C:\Users\donga\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
#Accesing the website
driver.get("https://archive.ics.uci.edu/ml/index.php ")
driver.maximize_window()
```

```python
dataset=[]
datatype=[]
task=[]
attributes=[]
instances=[]
num_attribute=[]
year=[]
```

```python
#Data
datasets=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/td[1]/table/tbody/tr/td[2] ')
data=[dt.text.split('\n') for dt in datasets]
dataset=[data[i][0] for i in range(len(data))]

#Data type
datatypes=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/td[2] ')
datatype=[da.text for da in datatypes]

#Task
default_task=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/td[3] ')
task=[t.text for t in default_task]

#attribute type
attributes=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/td[3] ')
attr=[at.text for at in attributes]

#Instances
instances_num=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/td[5] ')
instance=[ins.text for ins in instances_num]

#no. attribute
num_attribute=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/td[6] ')
num=[n.text for n in num_attribute]

#year
years=driver.find_elements(By.XPATH,'/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[2]/td[7] ')
year=[y.text for y in years]


UCI=pd.DataFrame({'Dataset':dataset[1:],'Datatypes':datatype[1:],'Task':task[1:],'Attributes':attr[1:],'Instances':instance[1:],
                  'No. of Attributes':num[1:], 'Year':year[1:]})
UCI
```
